### Converting  Date Time data into correct format:

```python
import datetime
# Udf for converting timestamp to time
def date_time_converter(time_stamp):
    sec_timestamp = time_stamp / 1000
    time_obj = datetime.datetime.fromtimestamp(sec_timestamp).time()
    return time_obj
```

```python
test_data['Recording timestamp'] = test_data['Recording timestamp'].apply(date_time_converter)
test_data['Recording duration'] = test_data['Recording duration'].apply(date_time_converter)
```

#### Imputing the pupil diameter left and right with mean reading of the recording

```python
test_data['Pupil diameter left'] = test_data.groupby(['Participant name', 'Recording name'])['Pupil diameter left'].transform(lambda x: x.fillna(x.mean()))
test_data['Pupil diameter right'] = test_data.groupby(['Participant name', 'Recording name'])['Pupil diameter right'].transform(lambda x: x.fillna(x.mean()))

test_data.head()
```
